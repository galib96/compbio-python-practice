```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
x = [0]
t = [0]

tend = 100

k = 1.5
gamma = 0.1

while t[-1] < tend:
    xi = x[-1]
    
    rates = [k, gamma*xi]
    rate_sum = sum(rates)
    
    tau = np.random.exponential(scale= 1/rate_sum)
    
    t.append(t[-1] + tau)
    
    r = random.uniform(0,1)
    
    # production event result
    if r*rate_sum > 0 and r*rate_sum < rates[0]:
        x.append(x[-1] + 1)
    elif r*rate_sum > rates[0] and r*rate_sum < rates[0] + rates[1]:
        x.append(x[-1] - 1)
    
```

```python
plt.plot(t,x)
plt.xlabel('Time')
plt.ylabel('mRNA quantity')
plt.show()
```

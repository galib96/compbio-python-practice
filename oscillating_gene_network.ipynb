```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
y0 = [0,0,0] # [G1, G2, G3]

t = np.linspace(0,200,100)

k_1 = 0.3
gamma_1 = 0.05
k_2 = 0.25
gamma_2 = 0.05
k_3 = 0.3
gamma_3 = 0.05

n = 8
c = 1

params = [k_1, gamma_1, k_2, gamma_2, k_3, gamma_3, n, c]
```

```python
def sim(variables, time, parameters):
    g1, g2, g3 = variables
    k_1, gamma_1, k_2, gamma_2, k_3, gamma_3, n, c = parameters
    
    dg1 = k_1*(c**n)/(c**n + g3**n) - gamma_1*g1
    dg2 = k_2*(g1**n)/(c**n + g1**n) - gamma_2*g2
    dg3 = k_3*(g2**n)/(c**n + g2**n) - gamma_3*g3
    
    return ([dg1, dg2, dg3])
```

```python
y = odeint(sim, y0, t, args=(params,))

f,ax = plt.subplots(1)

line1 = ax.plot(t, y[:,0], color='b', label='G1')
line2 = ax.plot(t, y[:,1], color='r', label='G2')
line3 = ax.plot(t, y[:,2], color='g', label='G3')

ax.set_ylabel('Number')
ax.set_xlabel('Time')
plt.legend()
plt.show()
```

```python
f, (ax1, ax2, ax3) = plt.subplots(3, sharex=True, sharey=False)

line1 = ax1.plot(t, y[:,0], color='b', label='G1')
line2 = ax2.plot(t, y[:,1], color='r', label='G2')
line3 = ax3.plot(t, y[:,2], color='g', label='G3')

ax1.set_ylabel('Number')
ax2.set_ylabel('Number')
ax3.set_ylabel('Number')
ax3.set_xlabel('Time')

ax1.legend()
ax2.legend()
ax3.legend()

plt.show()
```
